## Построение моделей и выбор лучшей модели

```python
def evaluate_model(model, X_test, Y_test):
    preds = model.predict(X_test)
    preds = preds.round(3) 
    mse = mean_squared_error(Y_test, preds)
    mae = mean_absolute_error(Y_test, preds)
    r2 = r2_score(Y_test, preds)
    return mse, mae, r2
```

```python
from sklearn.preprocessing import StandardScaler

def train_models():
    results = {}

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_val_scaled = scaler.transform(X_val)
    X_test_scaled = scaler.transform(X_test)

    catboost_model = CatBoostRegressor(verbose = 0, depth=4, random_state = 42)
    catboost_model.fit(X_train, Y_train, eval_set=(X_val, Y_val), early_stopping_rounds=50)
    results['CatBoost'] = evaluate_model(catboost_model, X_test, Y_test)
    
    lightgbm_model = LGBMRegressor(random_state=42, depth=4, early_stopping_rounds=50, verbose=-1)
    lightgbm_model.fit(X_train, Y_train, eval_set=[(X_val, Y_val)])
    results['LightGBM'] = evaluate_model(lightgbm_model, X_test, Y_test)
    
    # 3. XGBoost
    xgboost_model = XGBRegressor(random_state=42, depth=4)
    xgboost_model.fit(X_train, Y_train, eval_set=[(X_val, Y_val)], early_stopping_rounds=50, verbose=0)
    results['XGBoost'] = evaluate_model(xgboost_model, X_test, Y_test)

    rf_model = RandomForestRegressor(random_state=42)
    rf_model.fit(X_train, Y_train)
    results['RandomForest'] = evaluate_model(rf_model, X_test, Y_test)

    # ElasticNet Model
    elasticnet_model = ElasticNet(random_state=42)
    elasticnet_model.fit(X_train_scaled, Y_train)
    results['ElasticNet'] = evaluate_model(elasticnet_model, X_test_scaled, Y_test)


    return results, catboost_model, lightgbm_model, xgboost_model, rf_model, elasticnet_model

results, cbm, lgm, xbm, rfm, enm = train_models()

for model_name, (mse, mae, r2) in results.items():
    print(f"{model_name} -> MSE: {mse:.4f}, MAE: {mae:.4f}, R²: {r2:.4f}")
```

```python
def plot_feature_importance(importance, features, model_name):
    feature_importances_df = pd.DataFrame({'Feature': features, 'Importance': importance})
    feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)
    
    if feature_importances_df.empty:
        print(f"No feature importances found for {model_name}.")
        return
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=feature_importances_df)
    plt.title(f'Feature Importance for {model_name}')
    plt.show()
```

```python
cb_importances = cbm.get_feature_importance(prettified=True)
cb_importances_df = pd.DataFrame(cb_importances, columns=['Feature Id', 'Importances'])
plot_feature_importance(cb_importances_df['Importances'], cb_importances_df['Feature Id'], "CatBoost")

lg_importances = lgm.feature_importances_
plot_feature_importance(lg_importances, X_train.columns, "LightGBM")

xb_importances = xbm.feature_importances_
plot_feature_importance(xb_importances, X_train.columns, "XGBoost")

rf_importances = rfm.feature_importances_
plot_feature_importance(rf_importances, X_train.columns, "RandomForest")
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error

model = CatBoostRegressor(verbose=0, depth=6, random_state=42)

kf = KFold(n_splits=5, shuffle=True, random_state=42)

r2_scores = []
mae_scores = []
mape_scores = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    r2_scores.append(r2_score(y_test, y_pred))
    mae_scores.append(mean_absolute_error(y_test, y_pred))
    mape_scores.append(mean_absolute_percentage_error(y_test, y_pred))

print(f"Среднее значение R² на кросс-валидации: {np.mean(r2_scores):.4f}")
print(f"Среднее значение MAE на кросс-валидации: {np.mean(mae_scores):.4f}")
print(f"Среднее значение MAPE на кросс-валидации: {np.mean(mape_scores):.4f}")
```
